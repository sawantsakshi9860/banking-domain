```python
path=r'/Volumes/sakshi/dataframe/raw/day.csv'
df=spark.read.csv(path,header=True,inferSchema=True)
df.show()
df.display()

```

```python
from pyspark.sql.functions import col
df1=df.select(col('Numeric').isNull())
df1.show()
df1.display()
```

```python
df.createOrReplaceTempView("df")
df2=spark.sql("select count(*) from df where Numeric is null")
df2.display()
```

```python
from pyspark.sql.functions import when, col
df3 = df.withColumn("Numeric", when(col("Numeric") == 1, 0).otherwise(col("Numeric")))
df3.show()
df3.display()
```

```python
from pyspark.sql.functions import col
df1=df3.select(col('Numeric').isNull())
df1.show()
df1.display()
```

```python
from pyspark.sql.functions import lit
df1=df.withColumn('organization',lit("sachcloudy"))
df1.display()
df1.show()
```

```python
from pyspark.sql.functions import *
df2=df.withColumn("Process_run_date",current_date())\
    .withColumn("process_run_time",from_utc_timestamp(current_timestamp(),"Asia/Kolkata"))\
        .withColumn("process_run_year",year(current_date()))\
            .withColumn("process_run_month",month(current_date()))\
            .withColumn("process_run_day",dayofmonth(current_date()))\
                .withColumn("process_run",dayofweek(current_date()))\
                   .drop(col("process_run"))

df2.show()
df2.display()
```
